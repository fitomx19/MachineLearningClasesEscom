```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import csv
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

#importar el dataset
dataset = pd.read_csv("weatherAUS.csv")
#dataset

```

```python
dataset
```

```python
from cmath import nan


dataset.loc[dataset.RainToday == 'No','RainToday']='0'
dataset.RainToday.fillna(0, inplace=True)
dataset.loc[dataset.RainToday == 'Yes','RainToday']='1'

dataset.loc[dataset.RainTomorrow == 'No','RainTomorrow']='0'
dataset.RainTomorrow.fillna(0, inplace=True)
dataset.loc[dataset.RainTomorrow == 'Yes','RainTomorrow']='1' 

```

```python
x = dataset.iloc[:, -1:].values #todas menos la ultima columna
y  = dataset.iloc[:, :-1].values#solo la ultima columna
```

```python
## dividir el dataset en testing y en entrenamiento
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size= 0.2, random_state= 0,shuffle = True)
```

```python
#for x in dataset.RainTomorrow:
#    print(x)
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)
resultado = modelo.score(x_test, y_test)
print(resultado)
```

```python
from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier()
kfold_validacion = KFold(10) 
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, x, y, cv = kfold_validacion)
print(resultados)
resultados.mean() # para ver el promedio de los resultados 
```
